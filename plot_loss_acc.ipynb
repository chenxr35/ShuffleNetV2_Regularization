```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# exp_names = ["baseline", "data_augmentation", "L1", "L2", "L2_tune", "dropout"]
exp_names = {
    "baseline": "No regularization",
    "data_augmentation": "Data augmentation",
    "L1": "L1 0.001",
    "L2_tune": "L2 0.001",
    "L2": "L2 0.0001",
    "dropout": "Dropout"
}
```

```python
def plot_train_acc(exp_names):
    for exp_name in exp_names.keys():
        file_path = f"{exp_name}/fit_history.npz"
        save_data = np.load(file_path,allow_pickle=True)
        plt.plot(save_data['train_acc'])
    plt.title('Training accuracy')
    plt.ylabel('accuracy')
    plt.xlabel('epoch')
    plt.legend(exp_names.values(), loc='upper right')
    plt.grid(True)
    plt.savefig('training_accuracy.png')
    plt.show()
```

```python
plot_train_acc(exp_names)
```

```python
def plot_val_acc(exp_names):
    for exp_name in exp_names.keys():
        file_path = f"{exp_name}/fit_history.npz"
        save_data = np.load(file_path,allow_pickle=True)
        plt.plot(save_data['val_acc'])
    plt.title('Validation accuracy')
    plt.ylabel('accuracy')
    plt.xlabel('epoch')
    plt.legend(exp_names.values(), loc='upper right')
    plt.grid(True)
    plt.savefig('validation_accuracy.png')
    plt.show()
```

```python
plot_val_acc(exp_names)
```
