```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
exp_names = ["baseline", "data_augmentation", "L1"]
```

```python
def plot_train_acc(exp_names):
    for exp_name in exp_names:
        file_path = f"{exp_name}/fit_history.npz"
        save_data = np.load(file_path,allow_pickle=True)
        plt.plot(save_data['train_acc'])
    plt.title('Training accuracy')
    plt.ylabel('accuracy')
    plt.xlabel('epoch')
    plt.legend(exp_names, loc='upper right')
    plt.grid(True)
    plt.savefig('training_accuracy.png')
    plt.show()
```

```python
plot_train_acc(exp_names)
```

```python
def plot_val_acc(exp_names):
    for exp_name in exp_names:
        file_path = f"{exp_name}/fit_history.npz"
        save_data = np.load(file_path,allow_pickle=True)
        plt.plot(save_data['val_acc'])
    plt.title('Validation accuracy')
    plt.ylabel('accuracy')
    plt.xlabel('epoch')
    plt.legend(exp_names, loc='upper right')
    plt.grid(True)
    plt.savefig('validation_accuracy.png')
    plt.show()
```

```python
plot_val_acc(exp_names)
```

```python
def plot_train_loss(exp_names):
    for exp_name in exp_names:
        file_path = f"{exp_name}/fit_history.npz"
        save_data = np.load(file_path,allow_pickle=True)
        plt.plot(save_data['train_loss'])
    plt.title('Training loss')
    plt.ylabel('loss')
    plt.xlabel('epoch')
    plt.legend(exp_names, loc='upper right')
    plt.grid(True)
    plt.savefig('training_loss.png')
    plt.show()
```

```python
plot_train_loss(exp_names)
```
